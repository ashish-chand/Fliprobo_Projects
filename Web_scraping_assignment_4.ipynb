* Q- 5. Scrape the details of trending repositories on Github.com.
Url = https://github.com/
You have to find the following details:
A) Repository title
B) Repository description
C) Contributors count
D) Language used


Note: - From the home page you have to click on the trending option from Explore menu through code.

```python
# Activating the chrome browser
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(2)

# opening The website
url = "https://github.com/"
driver.get(url)
time.sleep(2)


#clicking the explore button
explore_btn = driver.find_element_by_xpath("/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/summary")
explore_btn.click()
time.sleep(1)

#clicking the trending button
trend_btn = driver.find_element_by_xpath("/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul[2]/li[3]/a")
trend_btn.click()
time.sleep(1)
```

```python
# Fetching urls of phones coming on 1st page
git_urls = []
urls = driver.find_elements_by_xpath('//h1[@class="h3 lh-condensed"]/a')
for url in urls:
    git_urls.append(url.get_attribute("href"))

#initialising empty list
git_df={}
git_df['Repositary title']=[]
git_df['Repsoitary description']=[]
git_df['Contributors count']=[]
git_df['Langauge used']=[]

# Scraping data from each url
for url in git_urls:
    driver.get(url)    # Saving url                                                   
    
    try:
        title = driver.find_element_by_xpath('//strong[@class="mr-2 flex-self-stretch"]')      # Extracting Title name from xpath
        git_df['Repositary title'].append(title.text)
    except NoSuchElementException:
        git_df['Repositary title'].append('-')
        
    try:
        desc = driver.find_element_by_xpath("//p[@class='f4 mt-3']")      # Extracting description from xpath
        git_df['Repsoitary description'].append(desc.text)
    except NoSuchElementException:
        git_df['Repsoitary description'].append('-')
        
    try:
        count=driver.find_element_by_xpath("//h2[@class='h4 mb-3']/a[contains(text(),'Contributors')]/span")
        git_df['Contributors count'].append(count.text)
    except NoSuchElementException:
        git_df['Contributors count'].append('-')
        
    try:
        lang = driver.find_element_by_xpath("//span[@class='color-text-primary text-bold mr-1']")      # Extracting language used from xpath
        git_df['Langauge used'].append(lang.text)
    except NoSuchElementException:
        git_df['Langauge used'].append('-')
        
df = pd.DataFrame(git_df)
df
```

```python
# Activating the chrome browser
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(2)

# opening The website
url = "https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare"
driver.get(url)
time.sleep(2)
```

```python
# Empty lists
novels_df = {}
novels_df['Book Name'] = []
novels_df['Author Name'] = []
novels_df['Volumes sold'] = []
novels_df['Publisher'] = []
novels_df['Genre'] = []

for i in range(0,101):
     # Book Name
    bname = driver.find_elements_by_xpath('//*[@id="table-cell-10943-{}-1"]'.format(i))
    for d in bname:
        novels_df['Book Name'].append(d.text)
        
     # Author Name
    a_name = driver.find_elements_by_xpath('//*[@id="table-cell-10943-{}-2"]'.format(i))
    for a in a_name:
        novels_df['Author Name'].append(a.text)
        
     # volume sold
    vsold = driver.find_elements_by_xpath('//*[@id="table-cell-10943-{}-3"]'.format(i))
    for vs in vsold:
        novels_df['Volumes sold'].append(vs.text)
        
     # Publisher
    pb = driver.find_elements_by_xpath('//*[@id="table-cell-10943-{}-4"]'.format(i))
    for p in pb:
        novels_df['Publisher'].append(p.text)
        
     # genre
    gen = driver.find_elements_by_xpath('//*[@id="table-cell-10943-{}-4"]'.format(i))
    for g in gen:
        novels_df['Genre'].append(g.text)
```
